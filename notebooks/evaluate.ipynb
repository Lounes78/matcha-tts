```python
import whisper
from jiwer import wer
import re
from IPython.display import Audio, display

# 1. Configuration
audio_path = 'output/generated_audio_synthes_example_original_model.wav'
# Texte original utilisé pour générer cet audio
original_text = "The Secret Service believed that it was very doubtful that any President would ride regularly in a vehicle with a fixed top, even though transparent."
```

```python
# display the audio
if os.path.exists(audio_path):
    display(Audio(audio_path, rate=22050))
```

```python
# 2. Chargement de Whisper Medium 
print("Chargement de Whisper Medium...")
asr_model = whisper.load_model("medium")
```

```python

# 3. Transcription de votre fichier sauvegardé
print(f"Transcription du fichier : {audio_path}")
result = asr_model.transcribe(audio_path)
predicted_text = result["text"]
```

```python

# 4. Normalisation pour une comparaison juste
def normalize(text):
    text = text.lower()
    return re.sub(r'[^\w\s]', '', text).strip()

ground_truth = normalize(original_text)
hypothesis = normalize(predicted_text)

# 5. Calcul du WER
error_rate = wer(ground_truth, hypothesis)

print(f"\n{'='*40}")
print(f"RÉSULTAT WER")
print(f"{'='*40}")
print(f"Texte prédit : {predicted_text}")
print(f"WER : {error_rate * 100:.2f} %")
print(f"{'='*40}")
```
